```python
# use df.page_text to produce an epub file with cleaned text
from reportlab.lib.pagesizes import LETTER
from reportlab.lib.units import inch
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, PageBreak
from reportlab.lib.styles import getSampleStyleSheet
from xml.sax.saxutils import escape
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont

pdfmetrics.registerFont(TTFont("DejaVuSans", "/Users/rogarcia/Library/Fonts/DejaVuSans.ttf"))
pdfmetrics.registerFont(TTFont("DejaVuSans-Bold", "/Users/rogarcia/Library/Fonts/DejaVuSans-Bold.ttf"))
pdfmetrics.registerFont(TTFont("DejaVuSans-Oblique", "/Users/rogarcia/Library/Fonts/DejaVuSans-Oblique.ttf"))
pdfmetrics.registerFont(TTFont("DejaVuSans-BoldOblique", "/Users/rogarcia/Library/Fonts/DejaVuSans-BoldOblique.ttf"))
pdfmetrics.registerFontFamily(
    "DejaVuSans",
    normal="DejaVuSans",
    bold="DejaVuSans-Bold",
    italic="DejaVuSans-Oblique",
    boldItalic="DejaVuSans-BoldOblique",
)
```

```python
def make_pdf_from_series(series, pdf_path, title="OCR Export"):
    doc = SimpleDocTemplate(
        pdf_path,
        pagesize=LETTER,
        leftMargin=inch,
        rightMargin=inch,
        topMargin=inch,
        bottomMargin=inch,
        title=title,
    )
    styles = getSampleStyleSheet()
    body_style = styles["BodyText"].clone("BodyTextUnicode")
    body_style.fontName = "DejaVuSans"
    header_style = styles["Heading4"].clone("Heading4Unicode")
    header_style.fontName = "DejaVuSans"
    story = []

    for i, text in enumerate(series, start=1):
        story.append(Paragraph(f"Page {i}", header_style))
        story.append(Spacer(1, 0.2 * inch))
        page_text = escape(str(text)).replace("\n", "<br/>")
        story.append(Paragraph(page_text, body_style))
        story.append(PageBreak())

    if story:
        story.pop()  # remove trailing PageBreak
    doc.build(story)
```

```python
make_pdf_from_series(df.page_text, "cleaned_ocr_output.pdf", title="Az Principles of Real Estate")
```
