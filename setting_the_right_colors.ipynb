```python
import flor
```

```python
df = flor.dataframe("page_color")
df
```

```python
# Pandas: select document_value, max(page) from df group by document_value
df.groupby('document_value').agg({'page': 'max'})
```

```python
# sum the max page values
df.groupby('document_value').agg({'page': 'max'}).sum()
```
